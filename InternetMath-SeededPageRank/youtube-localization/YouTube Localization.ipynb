```julia
using MatrixNetworks
```

```julia
A = readSMAT("com-youtube.ungraph.smat")
```

```julia
A = largest_component(A)[1]
```

```julia
d = sum(A,2)
#dind = findmax(d)[2] # this is the index
dind = 10000 # this is the index
```

```julia
@time x = personalized_pagerank(A,0.5,dind)
```

```julia
using Plots
pyplot()
```

```julia
plot(x) # this isn't "random" looking enough
```

```julia
srand(1)
p = randperm(size(A,1))
plot(x[p],grid=false,legend=false)
ylims!(0,0.05)
xlims!(1,size(A,1))
xlabel!("Index")
PyPlot.gca()[:spines]["top"][:set_visible](false) 
PyPlot.gca()[:spines]["right"][:set_visible](false) 

PyPlot.gca()[:yaxis][:set_ticks_position]("left")
PyPlot.gca()[:xaxis][:set_ticks_position]("bottom")

ylabel!("PageRank value")
```

```julia
PyPlot.gcf()[:set_size_inches](3.5,3.5)
savefig("youtube-localization-raw.pdf")
```

```julia
p = sortperm(x,rev=true)
plot(1-cumsum(x[p]),yaxis=(:log10, (1.e-4,1)),grid=false,legend=false)
PyPlot.gca()[:spines]["top"][:set_visible](false) 
PyPlot.gca()[:spines]["right"][:set_visible](false) 

PyPlot.gca()[:yaxis][:set_ticks_position]("left")
PyPlot.gca()[:xaxis][:set_ticks_position]("bottom")
xaxis!(:log10)
xlabel!("number of largest entries")
ylabel!("1-norm error")
```
